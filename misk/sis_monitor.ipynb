```python
import sys, time, pylab, numpy, datetime, pandas
import matplotlib.pyplot as plt
sys.path.append('/home/amigos/200Gmeasure/script/tools')
```

```python
d = numpy.loadtxt('/home/amigos/200Gmeasure/script/tools/TEST_hot_monitor_20180501-103810.txt')
t = [datetime.datetime.fromtimestamp(i) for i in d[:, 0]]
tt = [time.strftime('%H-%M-%S').format(_) for _ in t]
sisv = d[:,1]
sisi = d[:, 2]

fig = plt.figure(figsize=(14, 6))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.set_xlabel('JST', size=16)
ax1.set_ylabel('Voltage [mV]', size=16)
ax1.set_title('SISV Monitor @ C-lab', size=20)
ax1.tick_params(labelsize=14)
ax1.grid()
ax1.set_ylim(4.8, 4.95)
# ax1.set_xticklabels(tt, rotation=45)
ax1.plot(t, sisv, '.')


ax2.set_xlabel('JST', size=16)
ax2.set_ylabel('Current [uA]', size=16)
ax2.set_title('SISI Monitor @ C-lab', size=20)
ax2.tick_params(labelsize=14)
ax2.grid()
ax2.set_ylim(480, 495)
ax2.plot(t, sisi, '.')

plt.show()
```
