```python
pm = ML2437A.ml2437a(IP='192.168.100.161', GPIB=13)

now = datetime.datetime.now()
data = []

while 1:
    t = time.time()
    ret = pm.measure()
    data.append([t, ret])
    numpy.savetxt('TEST_hot_monitor', numpy.array(data), fmt=['%f', '%.3f'])
```
