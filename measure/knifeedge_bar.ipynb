```python
# 201804121

import os, sys, numpy, time, datetime

import time
start = time.time()

sys.path.append('/home/amigos/NASCORX_System-master/NASCORX_System/base')
sys.path.append('/home/amigos/NASCORX_System-master/NASCORX_System/device')
import Motor, ML2437A
import libMotor as lib

class pycolor(object):
    RED = '\033[31m'
    ENDC = '\033[0m'

slider = Motor.slider()
pm = ML2437A.ml2437a(IP='192.168.100.161', GPIB=13)
x = 0
y = 0
average = 1 # PowerMeter
length = 190
strk = 1
sleep_x = 0.3
sleep_y = 0.3

now = datetime.datetime.now()
os.mkdir('/home/amigos/sakasai/jupyter/beam_mesure/data/data_at_{0:%Y%m%d-%H%M%S}'.format(now))
os.chdir('/home/amigos/sakasai/jupyter/beam_mesure/data/data_at_{0:%Y%m%d-%H%M%S}'.format(now))

print(pycolor.RED + '\n\n' +
      '=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=\n'
      '   Start : Knifeedge Measurement  \n'
      '=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=\n'
      'x-axis : 0 [mm] - {0} [mm]\ny-axis : 0 [mm] - {0} [mm] \n'.format(length) +
      '\n\n' + pycolor.ENDC)

# initialize
slider.return_org(out=lib.RETURN_ORIGIN_XAXIS)
slider.return_org(out=lib.RETURN_ORIGIN_YAXIS)

data_x_cw = []
data_x_ccw = []
data_y_cw = []
data_y_ccw = []
# xdirection_measurement
## x_ccw
for i in range(length+1):
    ret_1x = []
    ret_2x = []
    for i in range(average):
        ret_= pm.measure() # system1
        ret_1x.append(ret_)
        ret_2x.append(0)
    ret_ave1 = round(numpy.mean(ret_1x), 6)
    ret_ave2 = round(numpy.mean(ret_2x), 6)
    data_x_ccw.append([time.time(), round(x, 1), ret_ave1, ret_ave2])
    numpy.savetxt('{0:%Y%m%d-%H%M%S}_x_ccw.csv'.format(now), numpy.array(data_x_ccw), delimiter=',', fmt=['%f', '%.0f', '%.6f', '%.6f'])
    slider.slide_ccw(strk=strk, axis='U')
    time.sleep(sleep_x)   
    msg = 'Axis : {0}\nStroke : {1} [mm]\nCoorValue : {0} = {2} [mm]\nDestinate : {0} = {3} [mm]'.format('X', strk, x, length)
    print('============'+'Knifeedge Measurement'+'============')
    print(msg)
    print('=========================================\n\n')
    x = x + strk
x = length
## x_cw
slider.slide_cw(strk=1, axis='U')
for i in range(length+1):
    ret_1x = []
    ret_2x = []
    for i in range(average):
        ret_= pm.measure() # system1
        ret_1x.append(ret_)
        ret_2x.append(0)
    ret_ave1 = round(numpy.mean(ret_1x), 6)
    ret_ave2 = round(numpy.mean(ret_2x), 6)
    data_x_cw.append([time.time(), round(x, 1), ret_ave1, ret_ave2])
    numpy.savetxt('{0:%Y%m%d-%H%M%S}_x_cw.csv'.format(now), numpy.array(data_x_cw), delimiter=',', fmt=['%f', '%.0f', '%.6f', '%.6f'])
    slider.slide_cw(strk=strk, axis='U')
    time.sleep(sleep_x)   
    msg = 'Axis : {0}\nStroke : {1} [mm]\nCoorValue : {0} = {2} [mm]\nDestinate : {0} = {3} '.format('X', strk, x, 'Origin!!')
    print('============'+'Knifeedge Measurement'+'============')
    print(msg)
    print('=========================================\n\n')
    x = x - strk
slider.return_org(out=lib.RETURN_ORIGIN_XAXIS) # return_origin
print(pycolor.RED + '\n\nx-axis Measurement finished!!!' + pycolor.ENDC)
print(pycolor.RED + 'x-axis Measurement finished!!!' + pycolor.ENDC)
print(pycolor.RED + 'x-axis Measurement finished!!!\n\n' + pycolor.ENDC)

a = input()

# ydirection_mesurement
## y_cw
for i in range(length+1):
    ret_1y = []
    ret_2y = []
    for i in range(average):
        ret_= pm.measure() # system1
        ret_1y.append(ret_)
        ret_2y.append(0)
    ret_ave1 = round(numpy.mean(ret_1y), 6)
    ret_ave2 = round(numpy.mean(ret_2y), 6)
    data_y_cw.append([time.time(), round(y, 1), ret_ave1, ret_ave2])
    numpy.savetxt('{0:%Y%m%d-%H%M%S}_y_cw.csv'.format(now), numpy.array(data_y_cw), delimiter=',', fmt=['%f', '%.0f', '%.6f', '%.6f'])
    slider.slide_cw(strk=strk, axis='Y')
    time.sleep(sleep_y)   
    msg = 'Axis : {0}\nStroke : {1} [mm]\nCoorValue : {0} = {2} [mm]\nDestinate : {0} = {3} '.format('Y', strk, y, length)
    print('============'+'Knifeedge Measurement'+'============')
    print(msg)
    print('=========================================\n\n')
    y = y + strk
y = length
## y_ccw
slider.slide_ccw(strk=1, axis='Y') # length  ni
for i in range(length+1):
    ret_1y = []
    ret_2y = []
    for i in range(average):
        ret_= pm.measure() # system1
        ret_1y.append(ret_)
        ret_2y.append(0)
    ret_ave1 = round(numpy.mean(ret_1y), 6)
    ret_ave2 = round(numpy.mean(ret_2y), 6)
    data_y_ccw.append([time.time(), round(y, 1), ret_ave1, ret_ave2])
    numpy.savetxt('{0:%Y%m%d-%H%M%S}_y_ccw.csv'.format(now), numpy.array(data_y_ccw), delimiter=',', fmt=['%f', '%.0f', '%.6f', '%.6f'])
    slider.slide_ccw(strk=strk, axis='Y')
    time.sleep(sleep_y)   
    msg = 'Axis : {0}\nStroke : {1} [mm]\nCoorValue : {0} = {2} [mm]\nDestinate : {0} = {3} [mm]'.format('Y', strk, y, 'Origin!!')
    print('============'+'Knifeedge Measurement'+'============')
    print(msg)
    print('=========================================\n\n')
    y = y - strk
slider.return_org(out=lib.RETURN_ORIGIN_YAXIS) # return_origin
print(pycolor.RED + '\n\ny-axis Measurement finished!!!' + pycolor.ENDC)
print(pycolor.RED + 'y-axis Measurement finished!!!' + pycolor.ENDC)
print(pycolor.RED + 'y-axis Measurement finished!!!' + pycolor.ENDC)
print(pycolor.RED + '\n\n'
      '=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=\n'
      '   Finish : Knifeedge Measurement  \n'
      '=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=\n'
      'x-axis : 0 [mm] - {0} [mm] \n'
      'y-axis : 0 [mm] - {0} [mm] \n'.format(length) + pycolor.ENDC)

slider.close_board()
```
